```python
import numpy as np
import math
import sys

def square(location, number_of_drones, drone_frame_size):
    
    if number_of_drones != 1:
        print ("wrong number of drones")
        sys.exit()
    
    new_loc = []
    x = location[0]
    y = location[1]
    if abs(x) > abs(y):
        if x > 0:
            new_loc = [x, y - 1]
        else:
            new_loc = [x, y + 1]
    else:
        if abs(x) < abs(y):
            if y > 0:
                new_loc = [x + 1, y]
            else:
                new_loc = [x - 1, y]
        else:# abs(x) = abs(y)
            if x > 0:
                if y > 0:
                    new_loc = [x + 1, y]
                else:
                    new_loc = [x - 1, y] 
            else:
                if x < 0:
                    if y > 0:
                        new_loc = [x + 1, y]
                    else:
                        new_loc = [x, y + 1]
                else: 
                    new_loc = [x + 1, y]
                    
                    
    new_drone_coords = []
    new_drone_coords.append([new_loc[0]*drone_frame_size[0], drone_frame_size[1]* new_loc[1]])
    
    return new_loc, new_drone_coords

def adapted_SQ(location, number_of_drones, drone_frame_size): #one drone

    if number_of_drones != 1:
        print ("wrong number of drones")
        sys.exit()
        
    
    new_loc = []
    x = location[0]
    y = location[1]
    
    if y < 0 and -y > abs(x):
        #zone 1
        print("zone 1")
        if y % 2 == 0:
            new_loc = [x + 1, y]
        else:
            new_loc = [x - 1, y]
                
    elif y > 0 and y > abs(2*x):
        #zone 2
        print("zone 2")
        temp_new_loc = []
        
        if y % 4 == 0:
            temp_new_loc = [x - 1, y]
        elif y % 4 ==1:
            temp_new_loc = [x - 1, y]
        elif y % 4 ==2:
            temp_new_loc = [x + 1, y]
        else:
            temp_new_loc = [x + 1, y]
        
        #check if outside zone 2, if so move up
        if temp_new_loc[1] < abs(2*temp_new_loc[0]):
            new_loc = [x, y+1]
        else:
            new_loc = temp_new_loc
        
    elif x > 0 and -x < y < 2*x:
        #zone 3
        print("zone 3")
        if x % 2 == 0:
            new_loc = [x, y + 1]
        else:
            new_loc = [x, y - 1]
                
    elif x < 0 and x < y < -2*x:
        #zone 4
        print("zone 4")
        if x % 2 == 0:
            new_loc = [x, y - 1]
        else:
            new_loc = [x, y + 1]
                
    elif y < 0 and x == -y:
        #coll 1 3
        print("coll 1 3")
        if x % 2 == 0:
            new_loc = [x, y + 1]
        else:
            new_loc = [x - 1, y]
        
    elif y < 0 and x == y:
        #coll 1 4
        print("coll 1 4")
        if x % 2 == 0:
            new_loc = [x + 1, y]
        else:
            new_loc = [x, y + 1]
        
    elif y > 0 and y == 2*x:
        #coll 2 3
        print("coll 2 3")
        if x % 2 == 0:
            new_loc = [x, y + 1]
        else:
            new_loc = [x, y - 1]
    
    elif y > 0 and y == -2*x:
        #coll 2 4
        print("coll 2 4")
        if x % 2 == 0:
            new_loc = [x, y - 1]
        else:
            new_loc = [x, y + 1]
    
    else:
        new_loc = [x, y + 1]
        
    new_drone_coords = []
    new_drone_coords.append([new_loc[0]*drone_frame_size[0], drone_frame_size[1]* new_loc[1]])
        
    return new_loc, new_drone_coords




def cone(location, conewidth, total_amount_of_drones, drone_frame_size): #0 is not included in drone number, conewidth in radians
    
    coneoffset = 2
    
    new_loc = []
    
    drone_path_angles = np.zeros(total_amount_of_drones)
    new_drone_coords = []

    x = location[0]
    y = location[1]
    
    drone_conewidth = conewidth / total_amount_of_drones
    
    for ii in range(total_amount_of_drones):
        drone_path_angles[ii] = total_amount_of_drones / 2 * drone_conewidth + 0.5 * drone_conewidth + ii * drone_conewidth
    
    if y % 2 == 0:
        new_loc = [x + 1, y]
        
    else:
        new_loc = [x - 1, y]
        
    
    
    
    if new_loc[1] != 0:
        
        
        drone_angle = abs(math.atan(new_loc[0] / new_loc[1]))    
        
        
        if drone_conewidth / 2 < drone_angle :
            new_loc = [x, y + 1]
    else:
        new_loc = [x, y + 1]
    
    for ii in range(total_amount_of_drones):
        newy = -drone_frame_size[0]*(new_loc[0] * math.cos(drone_path_angles[ii]) - new_loc[1] * math.sin(drone_path_angles[ii]))
        newx = drone_frame_size[1]*(new_loc[0] * math.sin(drone_path_angles[ii]) + new_loc[1] * math.cos(drone_path_angles[ii]))
        
        new_drone_coords.append([newx ,newy - coneoffset]) 
    
    return new_loc, new_drone_coords




def sweep_and_wait(location, sweepwidth, total_amount_of_drones, drone_frame_size):
    
    new_loc = []
    
    drone_path_width = sweepwidth / total_amount_of_drones
    new_drone_coords = []
    
    x = location[0]
    y = location[1]
    
    if y % 2 == 0:
        new_loc = [x + 1, y]
    else:
        new_loc = [x - 1, y]
        
    if abs(new_loc[0]) > drone_path_width / 2 and new_loc[0] > 0:
        new_loc = [x, y + 1]
    elif abs(new_loc[0]) > drone_path_width / 2 and new_loc[0] < 0:
        new_loc = [x, y - 1]
        
    for ii in range(total_amount_of_drones):
        
        new_drone_x = drone_frame_size[0]*(new_loc[0]-1/2 * (sweepwidth - drone_path_width) + ii*drone_path_width)
        new_drone_y = drone_frame_size[1]*new_loc[1]
        
        new_drone_coords.append([new_drone_x, new_drone_y])
        
    return new_loc, new_drone_coords
        

    
    

def target_movement(target_coords, stream_speed): #stream speed in cm/s
    
    heading = np.random.normal(loc=1/2 * np.pi, scale =  2)
    distance = np.random.uniform(stream_speed - 5, stream_speed + 5)
    
    new_target_coords = [target_coords[0] + math.cos(heading)*distance,target_coords[1] +  math.sin(heading)*distance]
    
    return new_target_coords








```

```python
drone_frame_size = [300, 240] #x, y
drone_speed = 10 #frame taken every n seconds, where n is the drone speed
drone_dispatch_time = 4*60 #in seconds

graphics = False
targetcolor = 'red'

stream_speed = 25 #cm per second
conewidth = 0.75 * np.pi
sweepwidth = 5
start_y = 1
number_of_drones = 1
number_of_trials = 10000
sim_runtime = 600
shrinkfactor = 0.1

succes_number = 0
succes_time = []
end_locations = []

if graphics == True: #initializing graphics
    import turtle
    
    target = turtle.Turtle()
    target.color(targetcolor)
    
    drone_turtles = []
    for ii in range(number_of_drones):
        drone_turtles.append(turtle.Turtle())
        drone_turtles[ii].goto(np.zeros(2))
    

for ii in range(number_of_trials): #looping for the amount of trials
    
    drone_location = [0,start_y]
    target_location = [0,0]
    found = False
    drone_location_list = []
    for jj in range(number_of_drones):
        drone_location_list.append(np.zeros(2))
    
    for jj in range(sim_runtime): #running a simulation
        
        target_location = target_movement(target_location, stream_speed)
        
        if drone_dispatch_time <= jj: #move the drone to the new location
            if jj % drone_speed == 0:
                
                drone_movement = sweep_and_wait(drone_location, sweepwidth, number_of_drones, drone_frame_size)
                #drone_movement = cone(drone_location, conewidth, number_of_drones, drone_frame_size)
                #drone_movement = adapted_SQ(drone_location, number_of_drones, drone_frame_size)
                #drone_movement = square(drone_location, number_of_drones, drone_frame_size)
                
                drone_location =  drone_movement[0]
                drone_location_list = drone_movement[1]
                
                for kk in range(number_of_drones): #check if each drone has found the target
                    if abs(drone_location_list[kk][0] - target_location[0]) < drone_frame_size[0]/2 and abs(drone_location_list[kk][1] - target_location[1]) < drone_frame_size[1]/2:
                        found = True    
                
                if graphics == True: #update the graphics for the drones
                    for kk in range(number_of_drones):
                        drone_turtles[kk].goto([x * shrinkfactor for x in drone_location_list[kk]])
                
        if graphics == True: #update the graphics for the target
            target.goto([x * shrinkfactor for x in target_location])
        
        if found == True: #end simulation if found, and log some results
            succes_number = succes_number + 1
            succes_time.append(jj)
            break
    
    end_locations.append(target_location) #log some more results
            
np.savetxt('results', succes_time)
```
